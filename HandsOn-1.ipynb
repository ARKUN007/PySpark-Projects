```python
from pyspark.sql.types import *

data = [
    ('S001','1/1/2025','ST01','P01','Grocery',2,100,'CARD'),
    ('S002','1/2/2025','ST01','P02','Grocery',None,200,'CASH'),  # '' replaced with None
    ('S003','3/5/2025','ST02','P03','Electronics',-1,1500,'UPI'),
    ('S004','3/10/2025','ST02','P04','Electronics',1,0,'CARD'),
    ('S005','','ST03','P05','Clothing',3,700,'UPI'),
    ('S006','15-04-2025','ST03','P06','Clothing',None,500,''),   # 'two' replaced with None
    ('S007','5/1/2025','ST01','P07','Grocery',5,300,'CARD'),
    ('S008','5/2/2025','ST02','P08','Grocery',0,250,'CASH'),
    ('S009','6/1/2025','ST03','P09','Electronics',2,-900,'UPI'),
    ('S010','6/5/2025','ST01','P10','Clothing',1,800,'')
]

schema = StructType([
    StructField('id',StringType(),True),
    StructField('date',StringType(),True),
    StructField('store',StringType(),True),
    StructField('product',StringType(),True),
    StructField('category',StringType(),True),
    StructField('quantity',IntegerType(),True),
    StructField('price',IntegerType(),True),
    StructField('payment',StringType(),True)
])

df = spark.createDataFrame(data=data,schema=schema)
df.show()
```

```python
df.select(["category","product"]).show()
df.count()
```

```python
df.show(5)
```

```python
from pyspark.sql.functions import col
df.select(
    col("id").alias("order_id"),
    col("date").alias("order_date"),
    col("store").alias("store_id"),
    col("product").alias("product_id"),
    col("category").alias("product_category"),
    col("quantity").alias("units"),
    col("price").alias("unit_price"),
    col("payment").alias("payment_method")
).show()
```

```python
from pyspark.sql.functions import lit
df = df.withColumn("tax", col('price') * 0.18)
df.show()
```
