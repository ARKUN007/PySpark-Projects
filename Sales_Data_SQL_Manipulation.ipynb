```python
from pyspark.sql.types import StringType, IntegerType, FloatType, StructType, StructField
from pyspark.sql.functions import *

data = [
    (1,'Phone',500,2,'2024-01-01'),
    (2,'Laptop',1000,1,'2024-01-02'),
    (3,'Tablet',300,3,'2024-01-03'),
    (4,'Phone',500,1,'2024-01-04'),
    (5,'Laptop',1000,2,'2024-01-05'),
    (6,'Tablet',300,1,'2024-01-06'),
    (7,'Phone',500,3,'2024-01-07'),
    (8,'Laptop',1000,1,'2024-01-08'),
    (9,'Tablet',300,2,'2024-01-09'),
    (10,'Phone',500,1,'2024-01-10'),
    (11,'Laptop',1000,3,'2024-01-11'),
    (12,'Tablet',300,1,'2024-01-12'),
    (13,'Phone',500,2,'2024-01-13'),
    (14,'Laptop',1000,1,'2024-01-14'),
    (15,'Tablet',300,3,'2024-01-15'),
    (16,'Phone',500,1,'2024-01-16'),
    (17,'Laptop',1000,2,'2024-01-17'),
    (18,'Tablet',300,1,'2024-01-18'),
    (19,'Phone',500,3,'2024-01-19'),
    (20,'Laptop',1000,1,'2024-01-20')
    ]

schema = StructType([
    StructField('id', IntegerType(), True),
    StructField('product', StringType(), True),
    StructField('price', IntegerType(), True),
    StructField('quantity', IntegerType(), True),
    StructField('date', StringType(), True)
])

df = spark.createDataFrame(data, schema)

df.show(truncate=False)
```

```python
df.createOrReplaceTempView("sales_view")
```

```python
df2 = spark.sql("select * from sales_view")
df2.show()
```

```python
%sql

select * from sales_view
```

```python
%sql
--highest sale
select * from sales_view
order by price * quantity desc
limit 1;
```

```python
%sql
--Second highest sale and revenue
select sv.*, sv.price * sv.quantity as revenue from sales_view sv
order by price * quantity desc
limit 1 offset 1
```
