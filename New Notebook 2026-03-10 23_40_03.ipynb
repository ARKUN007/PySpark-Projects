```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("pyspark_revision") \
    .getOrCreate()
```

```python
from pyspark.sql.types import StructType, StructField, StringType, IntegerType
data = [("John",1000),("Sam",2000), ("John",5000),("Sam",7000), ("avinash", None), ("Raushan", 9000), ("Sanam", None)]

schema = StructType([
    StructField("name", StringType()),
    StructField("salary", IntegerType())
])

df = spark.createDataFrame(data,schema)
df.show()
df.printSchema()
```

```python
df.select("name", "salary").show()
```

```python
df.filter(df.salary>1000).show()
```

```python
df.where(df.name =='John').show()
```

```python
df.show()
```

```python
df.select("name").distinct()
```

```python
from pyspark.sql.functions import col
df1 = df.withColumn("GST", df.salary*0.18)
df1.show()
```

```python
from pyspark.sql.functions import col
df1 = df.withColumn("GST", col("salary")*0.18)
df1.show()
```

```python
from pyspark.sql.functions import col
df1 = df1.withColumnRenamed("salary", "emp_salary")
df1.show()
```
